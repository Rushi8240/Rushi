```python
total_matches_played = data['team1'].value_counts().add(data['team2'].value_counts(), fill_value=0).astype(int)
total_matches_played
```

```python
total_won= data['match_winner'].value_counts()
total_won
```

```python
win_percentage = ((total_won / total_matches_played) * 100).sort_values(ascending=False).astype(int)
win_percentage
```

```python
team_performance = pd.DataFrame({
    'Total Matches Played': total_matches_played,
    'Total Matches Won': total_won,
    'Win Percentage (%)': win_percentage
}).sort_values(by='Win Percentage (%)', ascending=False)

team_performance
```

```python
px.bar(
    team_performance, x=team_performance.index, y=['Total Matches Played', 'Total Matches Won'],
    text_auto=True, barmode='group', title='Team Performance in IPL 2022',
    labels={'index':'IPL Team'},color_discrete_map={'Total Matches Played': 'lightblue', 'Total Matches Won': 'lightgreen'}
).update_layout(
    legend_title_text='Performance',yaxis_title='Match Counts Played & Won')
```

```python
import plotly.graph_objects as go

x,y1,y2 = team_performance.index,total_matches_played,total_won

fig = go.Figure(data=[
    go.Bar(name='Total Matches Played', x=x, y=y1,text=y1),
    go.Bar(name='Total Matches Won', x=x, y=y2,text=y2)
])
fig.update_layout(barmode='group',title='Team Performance in IPL 2022',
                  xaxis_title='IPL Team', yaxis_title='Counts Played & Won',
                  legend_title_text='Performance')
fig.show()
```

```python
px.bar(
    win_percentage, x=win_percentage.index, y=win_percentage,
    labels={'index': 'IPL Teams', 'y': 'Win Percentage'},
    title='Win Percentage by Each Team',text_auto=True, color=win_percentage
).update_layout(yaxis_ticksuffix='%')
```
